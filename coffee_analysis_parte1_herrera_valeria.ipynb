```python
import seaborn as sns
import matplotlib.pyplot as plt

df_articles['rating'] = pd.to_numeric(df_articles['rating'], errors='coerce')
df_articles['loc_country'] = df_articles['loc_country'].astype(str)

top_countries = df_articles['loc_country'].value_counts()
top_countries = top_countries[top_countries > 5].index
df_filtrado = df_articles[df_articles['loc_country'].isin(top_countries)]

plt.figure(figsize=(12, 6))
sns.boxplot(
    x='loc_country',
    y='rating',
    data=df_filtrado,
    palette='Pastel1'
)

plt.title('Distribución del rating por país de origen')
plt.xlabel('País de origen')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_articles,
    x='rating',
    y='100g_USD',
    hue='roast',
    palette='Set1'
)
plt.title('Relación entre rating y precio, según tipo de tueste')
plt.xlabel('Rating')
plt.ylabel('Precio (USD por 100g)')
plt.grid(True)
plt.show()
```

```python

top_countries = df_articles['loc_country'].value_counts().head(3).index
df_filtrado = df_articles[df_articles['loc_country'].isin(top_countries)]

plt.figure(figsize=(12, 6))
sns.boxplot(
    x='roast',
    y='100g_USD',
    hue='loc_country',
    data=df_filtrado,
    palette='Set2'
)
plt.title('Precio por tipo de tueste y país de origen')
plt.xlabel('Tipo de tueste')
plt.ylabel('Precio (USD por 100g)')
plt.legend(title='País')
plt.tight_layout()
plt.show()
```

```python

cols_numericas = ['rating', '100g_USD']
df_simple = df_articles[cols_numericas + ['roast']].dropna()

sns.pairplot(df_simple, hue='roast', palette='Set1')
plt.suptitle('Relaciones entre rating y precio por tipo de tueste', y=1.02)
plt.show()
```
